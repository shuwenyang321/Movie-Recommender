```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data from Github
df = pd.read_csv("https://github.com/ArinB/MSBA-CA-Data/raw/main/CA05/movies_recommendation_data.csv")
df
```

```python
#drop 'label' column
df = df.drop('Label',axis =1)
df
```

```python
movie = df['Movie Name']
movie
```

```python
movie_list = tuple, [[name] for name in df['Movie Name'].to_list()]
movie_list
```

```python
# features 
movie_features = df.loc[:,'IMDB Rating':'History']
movie_features
```

```python
#target
movie_target = df.loc[:,'Movie Name']
movie_target
```

```python
#build Knn model


from sklearn.neighbors import NearestNeighbors

#create feature vector for 'The Post'
PostFeature_vector = [7.2, 1, 1, 0, 0, 0, 0, 1]

k = 5
knn = NearestNeighbors(n_neighbors=k, algorithm="auto", metric="minkowski")
knn.fit(movie_features)

# Get the k nearest neighbors
distances, indices = knn.kneighbors([PostFeature_vector])

# for loop to print out the movie name
for movie_title in indices[0]:
    print(df.iloc[movie_title]['Movie Name'])
```

```python
import pickle
```

```python
#create a pickle file with Knn model 

data = {"model":knn, 
        "IMDB_Rating": np.array(df['IMDB Rating']), 
        "Biography": np.array(df['Biography']),
        "Drama" : np.array(df['Drama']),
        "Thriller" : np.array(df['Thriller']),
        "Comedy" : np.array(df['Comedy']),
        "Crime" : np.array(df['Crime']),
        "Mystery" : np.array(df['Mystery']),
        "History" : np.array(df['History'])
       }
with open('knn_model.pkl','wb') as file:
    pickle.dump(data,file)
```

```python
# read the pickle file 
with open('knn_model.pkl', 'rb') as file:
    data = pickle.load(file)
```
